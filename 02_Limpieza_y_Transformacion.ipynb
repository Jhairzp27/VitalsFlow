# Importación de librerías

```python
import pandas as pd
import numpy as np

# Cargar los datos crudos
df = pd.read_csv('healthcare_dataset.csv')
print(f"Filas originales: {df.shape[0]}")
```

# Limpieza de Texto (Estandarización)

```python
columnas_texto = ['Name', 'Doctor', 'Hospital', 'Medical Condition', 'Medication']
for col in columnas_texto:
    df[col] = df[col].astype(str).str.strip().str.title()
```

Arreglo nombres como 'ashLEy ERIcKSoN' a 'Ashley Erickson'

# Eliminación de Duplicados

En lugar de buscar filas 100% idénticas, borro registros si un paciente con el mismo nombre ingresó el mismo día, al mismo hospital y se le cobró exactamente lo mismo.
Esto elimina "falsos duplicados" con errores de tipeo en la edad.

```python
columnas_clave = ['Name', 'Date of Admission', 'Hospital', 'Billing Amount']
df = df.drop_duplicates(subset=columnas_clave, keep='first')
print(f"Filas después de deduplicación inteligente: {df.shape[0]}")
```

# Transformación Financiera (Facturación)

```python
df['Billing Amount'] = df['Billing Amount'].abs().round(2)
```

Los valores negativos son artefactos de la generación sintética.
Aplico valor absoluto y redondeo a 2 decimales (estándar financiero).

# Transformación Temporal (Fechas)

```python
df['Date of Admission'] = pd.to_datetime(df['Date of Admission'])
df['Discharge Date'] = pd.to_datetime(df['Discharge Date'])
```

# Ingeniería de Características (Feature Engineering para Q3)

```python
df['Days Hospitalized'] = (df['Discharge Date'] - df['Date of Admission']).dt.days
```

Calculamos el tiempo de estancia para responder la pregunta de negocio Q3.

# Verificación final

```python
print("\n--- REVISIÓN POST-LIMPIEZA ----\n")
print(f"Nulos restantes: {df.isnull().sum().sum()}")
print(f"Factura Mínima: ${df['Billing Amount'].min()}")
display(df[['Name', 'Age', 'Date of Admission', 'Billing Amount', 'Days Hospitalized']].head())
```

# Guardar el dataset procesado

```python
df.to_csv('healthcare_cleaned.csv', index=False)
print("\n¡Dataset limpio guardado exitosamente como 'healthcare_cleaned.csv'!")
```
